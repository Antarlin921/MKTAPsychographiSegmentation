```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Psychographic data 2.xlsx")
df.head()
```

```python
df2 = df.iloc[:, 1:15]
```

```python
df2.info()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

k_means = KMeans(n_clusters = 3,random_state=123, max_iter=1000)
model = k_means.fit_predict(df2)
```

```python
centers = k_means.cluster_centers_
```

```python
centers
```

```python
df2.mean()
```

```python
# Display the true centers
segments = ['Segment {}'.format(i) for i in range(0,len(centers))]
true_centers = pd.DataFrame(np.round(centers), columns = df2.keys())
true_centers.index = segments
true_centers
```

```python
clusters = k_means.labels_
df2['Clusters'] = clusters.tolist()
```

```python
df2['Clusters'].value_counts().sort_index()
```

```python
sns.countplot(x='Clusters', data=df2)
```
